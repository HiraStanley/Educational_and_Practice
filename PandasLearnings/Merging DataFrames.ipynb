```python
import pandas as pd
```

```python
customers = pd.read_csv("Restaurant - Customers.csv")
customers.head()
```

```python
foods = pd.read_csv("Restaurant - Foods.csv")
foods.head()
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week1.head()
```

```python
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
week2.head()
```

# Concat Function

```python
len(week1)
```

```python
len(week2)
```

```python
sales = pd.concat([week1, week2], ignore_index=True)
```

```python
sales
```

```python
pd.concat([week1, week2]).sort_index()
```

```python
pd.concat([week1, week2], keys=["Week 1", "Week 2"])
```

# Concat with Different Column Names

```python
df1 = pd.DataFrame([1, 2, 3], columns=["A"])
df1
```

```python
df2 = pd.DataFrame([4, 5, 6], columns=["B"])
df2
```

```python
pd.concat([df1, df2])
```

# Concat on Column Axis

```python
pd.concat([df1, df2], axis="columns")
```

# Left Joins

```python
week1.head()
```

```python
foods.head()
```

```python
week1.merge(foods, "left", on="Food ID")
```

# left_on and right_on Parameters

```python
customers.head()
```

```python
week2.head()
```

```python
week2.merge(customers, how="left", left_on="Customer ID", right_on="ID")
```

```python
week2.merge(customers, how="left", left_on="Customer ID", right_on="ID").drop("ID", axis="columns")
```

# Inner Joins

```python
week1.merge(week2, how="inner", on="Customer ID", suffixes=[" - Week 1", " - Week 2"])
```

# Inner Joins with a List

```python
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"])
```

# Full Outer Join

```python
week1.merge(week2, how="outer", on="Customer ID")
```

```python
week1.merge(week2, how="outer", on="Customer ID", indicator=True)
```

```python
merged = week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - Week 1", " - Week 2"], indicator=True)
merged
```

```python
merged["_merge"].value_counts()
```

```python
merged[merged["_merge"].isin(["left_only", "right_only"])]
```

```python
customers.head()
```

```python
week1.merge(customers, how="left", left_on="Customer ID", right_index=True)
```
