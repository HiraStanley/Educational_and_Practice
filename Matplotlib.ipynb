```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(6,6))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared',color='red', lw=2,alpha=0.5, marker='*',ms=10)
ax.plot(x,x**3, label = 'X Cubed', color='purple', lw=2, ls='--', marker='o', ms=10, mfc='red', mew=3, mec='green')

ax.legend(loc=0)
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure(figsize=(6,6))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,3])
ax.set_ylim([0,5])
```
