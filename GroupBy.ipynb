```python
import pandas as pd
```

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
fortune.head()
```

```python
fortune.nunique()
```

# Pivot Method

```python
fortune.pivot_table(values="Company", index=["Sector", "Industry"], aggfunc="count")
```

```python
fortune[fortune["Sector"] == "Retailing"]["Revenue"].sum()
```

# Methods on GroupBy Object

```python
sectors["Revenue"].sum()
sectors["Employees"].min()
sectors["Profits"].max()
sectors["Employees"].mean()
```

```python
sectors[["Revenue", "Profits"]].sum()
```

# Grouping by Multiple Columns

```python
sectors = fortune.groupby(["Sector", "Industry"])
```

```python
sectors.size()
```

```python
sectors["Revenue"].sum()
```

# The agg method

```python
sectors = fortune.groupby("Sector")
```

```python
# same as doing sectors["Revenue"].sum() but can do multiple aggs
sectors.agg({"Revenue": "sum", "Profits":"max", "Employees":"mean"})
```

# Apply Method

```python
# Find the two companies in each sector with the most employees
```

```python
fortune.nlargest(2, "Employees")
```

```python
def top_two_companies_by_employee_count(sector):
    return sector.nlargest(2, "Employees")

sectors.apply(top_two_companies_by_employee_count)
```

```python
states = fortune.groupby("State")
```

```python
def top_company_by_profits(states):
    return states.nlargest(1, "Profits")

states.apply(top_company_by_profits)
```

```python
def top_company_by_employees(states):
    return states.nlargest(1, "Employees")

states.apply(top_company_by_employees)
```
